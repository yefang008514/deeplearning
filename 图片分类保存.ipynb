```python
import shutil
os.mkdir('./train/')
os.mkdir('./train/CBB')
os.mkdir('./train/CBSD')
os.mkdir('./train/CGM')
os.mkdir('./train/CMD')
os.mkdir('./train/H')
os.mkdir('./train/AL')
import pandas as pd
train_table=pd.read_csv('train.csv')
img_name=[]
for i in range(0,6):
  img_name.append(train_table[train_table['label']==i]['image_id'])

for j in range(0,6):#读取图片路径并保存
  for k in img_name[j]:
    path = './train_img/'+k
    if j==0: 
      path_to = './train/CBB/'+k
    elif j==1:
      path_to = './train/CBSD/'+k
    elif j==2:
      path_to = './train/CGM/'+k
    elif j==3:
      path_to = './train/CMD/'+k
    elif j==4:
      path_to = './train/H/'+k
    elif j==5:
      path_to = './train/AL/'+k
    # print(path,j,path_to) 
    shutil.copy(os.path.abspath(path),os.path.abspath(path_to))
```
